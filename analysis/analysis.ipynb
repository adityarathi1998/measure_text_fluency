### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

### Read Data

```python
BASE_PATH = "../output/"
```

```python
file_path = BASE_PATH + "output_time.txt"
```

```python
data=pd.read_csv(file_path, sep="\t")
data.columns = ["bleu score", "rouge score", "meteor score"]
```

```python
data.head()
```

```python
data.dtypes
```

```python
data_average = data.tail(6)
data.drop(data.tail(6).index,inplace=True) # drop last n rows
data = data.astype(np.float32)
```

### Plots


```python
rows = data.shape[0]
rows
```

```python
plt.figure(figsize=(10,10))
plt.plot(range(rows), data["bleu score"].sort_values(), label = "bleu score")
plt.plot(range(rows), data["rouge score"].sort_values(), label = "rouge score")
plt.plot(range(rows), data["meteor score"].sort_values(), label = "meteor score")
plt.legend()

plt.xlabel("Sentence number")
plt.ylabel("Scores")
plt.title('Fluency Score')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.kdeplot(data["bleu score"], shade=True, label = "bleu score")
sns.kdeplot(data["rouge score"], shade=True, label = "rouge score")
sns.kdeplot(data["meteor score"], shade=True, label = "meteor score")
plt.legend()
plt.xlabel("scores")
plt.show()
```
